```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculate the requested summaries and load them into SummaryData
summary_data = [{"Number of Unique Items" : len(purchase_data.groupby("Item ID")),
                "Average Price" : round(purchase_data["Price"].mean(), 2), 
                "Number of Purchases" : len(purchase_data.groupby("Purchase ID")),
                "Total Revenue" : purchase_data["Price"].sum()}]

summary_df= pd.DataFrame(summary_data)

# Change the order of the columns
summary_df = summary_df[["Number of Unique Items","Average Price","Number of Purchases","Total Revenue"]]
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Grab just the gender data and unique player names from the data and group it by gender
gender_data = purchase_data.groupby(["Gender"])["SN"].nunique()
# Convert to a dataframe summarized by gender with one column, rename that column, calculated percentage column and add it
gender_df = pd.DataFrame(gender_data)
gender_df.columns = ["Total Players"]
gender_df["Percentage of Players"] = round(gender_df["Total Players"]/gender_df["Total Players"].sum()*100, 2)
# Print out gender data sorted by number of players
gender_df.sort_values(by = "Total Players", ascending=False)
```

Conclusion - Boys and Men like this game a lot more than Girls and Women



## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Generate a puchase summary grouped by gender, with the aggregations needed for the desired table
purchase_summary = purchase_data.groupby(["Gender"]).agg({"Price" : ["mean", "sum"], "Purchase ID" : "count", "SN" : "nunique"})
```

```python
purchase_df = pd.DataFrame(purchase_summary)
purchase_df.head()
```

```python
# Convert the purchase summary data back to a dataframe organized by gender, rename the columns, add the avg purchase and output
purchase_df = pd.DataFrame(purchase_summary)
purchase_df.columns = ["Average Purchase Price", "Total Purchase Value", "Purchase Count", "Player Count"]
purchase_df["Avg Total Purchase per Person"] = round(purchase_df["Total Purchase Value"]/purchase_summary["Player Count"],2)
purchase_df = purchase_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Player Count", "Avg Total Purchase per Person"]]
purchase_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
print(purchase_data["Age"].min(), purchase_data["Age"].max())
```

```python
# Generate a list of age bins. Min age was 7, max was 45. Cut the data by those bins
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
age_hist = pd.cut(purchase_data["Age"], age_bins)
# Copy the source data because I want to preserve the original data. Add the histogram category column to the new data frame
age_df = purchase_data.copy()
age_df["age_hist"] = age_hist
# Group the data by the histogram categories, with all tne needed aggregations for all requested Age analysis
age_summary = age_df.groupby(["age_hist"]).agg({"Price" : ["mean", "sum"], "Purchase ID" : "count", "SN" : "nunique"})
# Build a summary dataframe indexed by the histogram categories, add the calculated columns and get columns in logical order
age_summary_df = pd.DataFrame(age_summary)
age_summary_df.columns = ["Average Purchase Price", "Total Purchase Value", "Purchase Count","Player Count"]
age_summary_df["Avg Total Purchase per Person"] = round(age_summary_df["Total Purchase Value"]/age_summary_df["Player Count"], 2)
age_summary_df["Percentage of Players"] = round(age_summary_df["Player Count"]/age_summary_df["Player Count"].sum()*100, 2)
age_summary_df = age_summary_df[["Player Count", "Percentage of Players",  "Purchase Count", "Total Purchase Value", "Average Purchase Price", "Avg Total Purchase per Person"]]
age_summary_df
```

```python
# Select the columns of interest from the dataframe and output the new dataframe for Age Demographics
player_demographics_df = age_summary_df[["Player Count", "Percentage of Players"]]
player_demographics_df
```

Conclusion: This game is mostly played by Millenials (and younger)

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders_summary = purchase_data.groupby("SN").agg({"Purchase ID" : "count", "Price" : "sum"})
top_spenders_df = pd.DataFrame(top_spenders_summary)
top_spenders_df.columns = ["Purchase Count", "Total Purchase Value"]
top_spenders_df["Average Purchase Price"] = round(top_spenders_df["Total Purchase Value"] / top_spenders_df["Purchase Count"], 2)
top_spenders_df = top_spenders_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]

top_spenders_df.sort_values("Total Purchase Value", ascending=False).head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items_df = purchase_data[["Item ID", "Item Name", "Price"]]
items_summary = items_df.groupby(["Item ID", "Item Name"]).agg({"Item ID" : "count", "Price" : ["mean", "sum"]})
items_summary_df = pd.DataFrame(items_summary)
items_summary_df.columns = ["Purchase Count", "Purchase Price", "Total Purchases"]
items_summary_df.sort_values("Purchase Count", ascending = False).head(10)
```

Conclusion: Purchase price of items does not have much impact on purchases. Price is perhaps not reflective of value to player.
